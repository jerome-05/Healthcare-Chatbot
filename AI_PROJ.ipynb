```python
# Install necessary libraries
!pip install transformers gradio serpapi

import gradio as gr
from transformers import pipeline
import requests
import json

# Load pre-trained model from Hugging Face
nlp = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Function to fetch context using SerpAPI
def get_context(query):
    api_key = "60a291827fc073ecfeec0d25a7a5a81071c19518320a970538918bb3ad4daaf3"  # Replace with your actual SerpAPI key
    search_url = f"https://serpapi.com/search.json?q={query}&api_key={api_key}"

    response = requests.get(search_url)

    if response.status_code == 200:
        search_results = response.json()
        if "organic_results" in search_results and len(search_results["organic_results"]) > 0:
            # Get the top result snippet as context
            context = search_results["organic_results"][0].get("snippet", "Context not available")
            return context
        else:
            return "No context found from search."
    else:
        return "Error fetching search results."

# Function to use QA pipeline for answering questions
def healthcare_bot(question):
    # Fetch context from web using SerpAPI
    context = get_context(question)

    # Apply the model if context is found
    if context and context != "No context found from search.":
        result = nlp(question=question, context=context)
        return f"Answer: {result['answer']}\n\nContext: {context}"
    else:
        return context

# Launch the Gradio interface
iface = gr.Interface(fn=healthcare_bot, inputs="text", outputs="text", title="Healthcare Chatbot",
                     description="Ask healthcare-related questions and get answers from web-based context.")
iface.launch()
```
